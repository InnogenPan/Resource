```python
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
img = cv2.imread("/home/wangfeng02/cat.jpg")
print("image shape: {}".format(img.shape))
plt.imshow(img)
```

```python
from megengine.data import transform as T
transforms=[
    T.ShortestEdgeResize(
        min_size=800,
        max_size=1333,
        sample_style="choice",
    ),
    T.RandomHorizontalFlip(prob=0.5),
    T.ToMode("CHW"),
]

transform_img = img
print("image shape before transform: {}".format(img.shape))
for t in transforms:
    print(t)
    transform_img = t.apply(transform_img)
    print(transform_img.shape)
    if isinstance(t, T.RandomHorizontalFlip):
        plt.imshow(transform_img)
```
